# Group By and Aggregate 

```python
df=pd.read_csv(r"C:\Users\PC\Desktop\DATA ANALYSIS\Pandas Tutorials\Flavors.csv")

df
```

```python
group_by_flavours=df.groupby('Base Flavor')
```

```python
group_by_flavours.mean()
```

```python
group_by_flavours.sum()
```

```python
group_by_flavours.max()
```

```python
group_by_flavours.agg({'Flavor Rating':['mean','max','sum','count'],'Texture Rating':['mean','max','sum','count']})
```

```python
df.groupby(['Base Flavor','Liked']).max()
```

```python
df.groupby(['Base Flavor','Flavor']).agg({'Flavor Rating':['mean','max','sum','count'],'Texture Rating':['mean','max','sum','count']})
```
